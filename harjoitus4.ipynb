### 1. Autodatan yhdistäminen

Tee funktio autot(url1, url2), joka tekee järjestyksessä seuraavaa:
* lataa datat DataFrameihin url1:sta ja url2:sta, joissa erotinmerkkinä on kaksoispiste (';'),

```python
#import pandas as pd
#import numpy as np

#url_a1='https://student.labranet.jamk.fi/~huoptu/datananal_s20/kerta4/harjoitukset/autotA1.csv'
#url_a2='https://student.labranet.jamk.fi/~huoptu/datananal_s20/kerta4/harjoitukset/autotA2.csv'
#url_b1='https://student.labranet.jamk.fi/~huoptu/datananal_s20/kerta4/harjoitukset/autotB1.csv'
#url_b2='https://student.labranet.jamk.fi/~huoptu/datananal_s20/kerta4/harjoitukset/autotB2.csv'

def autot(url_a1, url_a2, url_b1, url_b2):
    # Kirjoita vastauksesi tähän
    # ...

    df_a1 = pd.read_csv(url_a1, sep=";")
    #print(df_a1.dtypes)
    #df_a1 #1099 rows x 16 columns

    df_a2 = pd.read_csv(url_a2, sep=";")
    #print(df_a2.dtypes)
    #df_a2 #1100 rows x 16 columns

    df_b1 = pd.read_csv(url_b1, sep=";")
    #print(df_b1.dtypes)
    #print("this is df_b1")
    #df_b1 #1099 rows x 18 columns

    #print('\n---\n')
    df_b2 = pd.read_csv(url_b2, sep=";")
    #print(df_b2.dtypes)
    #print("this is df_b2")
    #df_b2 #1101 rows x 18 columns

    #print('\n---\n')
    df_b1['Ernedc (g/km)'] = df_b1['Ernedc (g/km)'].astype(object)
    #print(df_b1.dtypes)
    #print("this is dtype converted df_b1")

    #print('\n---\n')
    df_b2['ec (cm3)'] = df_b2['ec (cm3)'].astype(float)
    #print(df_b2.dtypes)
    #print("this is dtype converted df_b2")

    #nämä toimivat anacondassa, mutta vanhempi versio halusi concat float arvoille
    #dfa_combined = pd.merge(df_a1, df_a2, how="outer")
    #dfb_combined = pd.merge(df_b1, df_b2, how="outer")
    
    dfa_combined = pd.concat([df_a1,df_a2],ignore_index=True)
    dfb_combined = pd.concat([df_b1,df_b2],ignore_index=True)

    df2 = pd.merge(dfa_combined, dfb_combined, on='ID')
    #df2 #2199 rows × 33 columns

    df2['ep (KW)'] = df2['ep (KW)'].replace(np.nan, 0)

    df2 = pd.DataFrame(df2.groupby('Mh')['ep (KW)'].mean().reset_index())

    df2 = df2.sort_values('ep (KW)', ascending=False)

    df2 = df2.head()
    tulos = df2
    #tulos

    # Älä kirjoita muita print komentoja ennen tulos muuttujan tulostusta.
    print(tulos)
    
# Testaa funktiotasi:
url_a1='https://student.labranet.jamk.fi/~huoptu/datananal_s20/kerta4/harjoitukset/autotA1.csv'
url_a2='https://student.labranet.jamk.fi/~huoptu/datananal_s20/kerta4/harjoitukset/autotA2.csv'
url_b1='https://student.labranet.jamk.fi/~huoptu/datananal_s20/kerta4/harjoitukset/autotB1.csv'
url_b2='https://student.labranet.jamk.fi/~huoptu/datananal_s20/kerta4/harjoitukset/autotB2.csv'

autot(url_a1, url_a2, url_b1, url_b2)
```

```python
#import pandas as pd
#import numpy as np

#url_a1='https://student.labranet.jamk.fi/~huoptu/datananal_s20/kerta4/harjoitukset/autotA1.csv'
#url_a2='https://student.labranet.jamk.fi/~huoptu/datananal_s20/kerta4/harjoitukset/autotA2.csv'
#url_b1='https://student.labranet.jamk.fi/~huoptu/datananal_s20/kerta4/harjoitukset/autotB1.csv'
#url_b2='https://student.labranet.jamk.fi/~huoptu/datananal_s20/kerta4/harjoitukset/autotB2.csv'

def autot(url_a1, url_a2, url_b1, url_b2):
    # Kirjoita vastauksesi tähän
    # ...
    df3_a1 = pd.read_csv(url_a1, sep=";")
    df3_a2 = pd.read_csv(url_a2, sep=";")
    df3_b1 = pd.read_csv(url_b1, sep=";")
    df3_b2 = pd.read_csv(url_b2, sep=";")

    df3_b1['Ernedc (g/km)'] = df3_b1['Ernedc (g/km)'].astype(object)

    df3_b2['ec (cm3)'] = df3_b2['ec (cm3)'].astype(float)
    
    #nämä toimivat anacondassa, mutta vanhempi versio halusi concat float arvoille
    #dfa3_combined = pd.merge(df3_a1, df3_a2, how="outer")
    #dfb3_combined = pd.merge(df3_b1, df3_b2, how="outer")
    
    dfa3_combined = pd.concat([df3_a1,df3_a2],ignore_index=True)
    dfb3_combined = pd.concat([df3_b1,df3_b2],ignore_index=True)

    df3 = pd.merge(dfa3_combined, dfb3_combined, on='ID')

    df3['Ft'] = df3['Ft'].replace('DIESEL', 'Diesel')
    df3['Ft'] = df3['Ft'].replace('PETROL', 'Petrol')

    df4 = df3
    df4 = pd.crosstab(df4['Mh'], df4['Ft'], df4['Enedc (g/km)'], aggfunc='mean').applymap("{:.1f}".format) 

    df4 = df4.replace(np.nan, 'nan')

    df5 = df4[['Diesel', 'Petrol']]
    df5 = df5.head()
    tulos = df5
    #tulos

    # Älä kirjoita muita print komentoja ennen tulos muuttujan tulostusta.
    print(tulos)
    
# Testaa funktiotasi:
url_a1='https://student.labranet.jamk.fi/~huoptu/datananal_s20/kerta4/harjoitukset/autotA1.csv'
url_a2='https://student.labranet.jamk.fi/~huoptu/datananal_s20/kerta4/harjoitukset/autotA2.csv'
url_b1='https://student.labranet.jamk.fi/~huoptu/datananal_s20/kerta4/harjoitukset/autotB1.csv'
url_b2='https://student.labranet.jamk.fi/~huoptu/datananal_s20/kerta4/harjoitukset/autotB2.csv'

autot(url_a1, url_a2, url_b1, url_b2)
```

```python
#import pandas as pd
#import numpy as np

#url_a1='https://student.labranet.jamk.fi/~huoptu/datananal_s20/kerta4/harjoitukset/autotA1.csv'
#url_a2='https://student.labranet.jamk.fi/~huoptu/datananal_s20/kerta4/harjoitukset/autotA2.csv'
#url_b1='https://student.labranet.jamk.fi/~huoptu/datananal_s20/kerta4/harjoitukset/autotB1.csv'
#url_b2='https://student.labranet.jamk.fi/~huoptu/datananal_s20/kerta4/harjoitukset/autotB2.csv'

def autot(url_a1, url_a2, url_b1, url_b2):
    # Kirjoita vastauksesi tähän
    # ...
    df8_a1 = pd.read_csv(url_a1, sep=";")
    df8_a2 = pd.read_csv(url_a2, sep=";")
    df8_b1 = pd.read_csv(url_b1, sep=";")
    df8_b2 = pd.read_csv(url_b2, sep=";")

    df8_b1['Ernedc (g/km)'] = df8_b1['Ernedc (g/km)'].astype(object)
    df8_b2['ec (cm3)'] = df8_b2['ec (cm3)'].astype(float)

    #nämä toimivat anacondassa, mutta vanhempi versio halusi concat float arvoille
    #dfa8_combined = pd.merge(df8_a1, df8_a2, how="outer")
    #dfb8_combined = pd.merge(df8_b1, df8_b2, how="outer")
    
    dfa8_combined = pd.concat([df8_a1,df8_a2],ignore_index=True)
    dfb8_combined = pd.concat([df8_b1,df8_b2],ignore_index=True)

    df8 = pd.merge(dfa8_combined, dfb8_combined, on='ID')

    df8['Ft'] = df8['Ft'].replace('DIESEL', 'Diesel')
    df8['Ft'] = df8['Ft'].replace('PETROL', 'Petrol')

    df9 = df8

    df9 = df9[['Enedc (g/km)', 'MS', 'Ft']]

    df9 = df9.rename(columns={'Enedc (g/km)' : 'Avg emissions (g/km)'})
    df9 = df9.rename(columns={'MS' : 'Member State'})
    df9 = df9.rename(columns={'Ft' : 'Fuel type'})

    df9 = pd.crosstab(df9['Member State'],df9['Fuel type'], values=df9['Avg emissions (g/km)'], aggfunc=np.mean, dropna=False)

    df9 = df9.astype(float)
    df9['D-P'] = df9['Diesel'] - df9['Petrol']
    df9 = df9.sort_values('D-P', ascending=False)
    df9 = df9.applymap("{:.0f}".format)

    df9.columns = pd.MultiIndex.from_product([['Avg emissions (g/km)'], df9.columns])

    df9 = df9.drop(['Electric', 'NG'], axis=1, level=1)
    df9 = df9.head()
    tulos = df9
    #tulos

    # Älä kirjoita muita print komentoja ennen tulos muuttujan tulostusta.
    print(tulos)
    
# Testaa funktiotasi:
url_a1='https://student.labranet.jamk.fi/~huoptu/datananal_s20/kerta4/harjoitukset/autotA1.csv'
url_a2='https://student.labranet.jamk.fi/~huoptu/datananal_s20/kerta4/harjoitukset/autotA2.csv'
url_b1='https://student.labranet.jamk.fi/~huoptu/datananal_s20/kerta4/harjoitukset/autotB1.csv'
url_b2='https://student.labranet.jamk.fi/~huoptu/datananal_s20/kerta4/harjoitukset/autotB2.csv'

autot(url_a1, url_a2, url_b1, url_b2)
```
